```python
def create_bed_file(file_name):
  counter = 1
  with open(file_name + ".bed6", 'w') as genome:
    with open(file_name + "_tss.bed4", 'w') as tss:
      with open(file_name + "_all.bed4", 'w') as all:
        with open(f'{file_name}_feature_table.txt') as file:
          line = file.readline()
          line = file.readline()

          while (line):
            tokens1 = line.split('\t')
            line = file.readline()
            tokens2 = line.split('\t')

            start = tokens1[7]
            end = tokens1[8]
            strand = tokens1[9]
            if int(start) - int(end) < 50000:
              name = tokens2[10]
              if not name:
                name = f'empty_id_{counter}'
                counter += 1
              genome.write(f'chr1\t{start}\t{end}\t{name}\t1000\t{strand}\n')
              start_original = start
              end_original = end
              if strand == '+':
                start = (int(start) - 100 if int(start) - 100 > 0 else 0)
                end = int(start) + 200
                start_original = start
              else:
                end = int(end) + 100
                start = (int(end) - 200 if int(end) - 200 > 0 else 0)
                end_original = end
              tss.write(f'chr1\t{start}\t{end}\t{name}\n')
              all.write(f'chr1\t{start_original}\t{end_original}\t{name}\n')

              line = file.readline()
        
for name in file_names:
  create_bed_file(name)
```

```python
import os
import subprocess
import tempfile
import re
from pathlib import Path
from subprocess import DEVNULL, PIPE

import pandas as pd

ZH_EXECUTABLE = Path("/content/zhunt3")
assert ZH_EXECUTABLE.is_file()


def zhunt(query: str, windowsize: int = 6, minsize: int = 3, maxsize: int = 6):
    query = re.sub('[U,W,S,M,K,R,Y,B,D,H,V]', 'N', query)
    assert set(query).issubset({"A", "C", "G", "T", "N"})
    fd, temp = tempfile.mkstemp()
    os.close(fd)
    with open(temp, 'w') as stream:
        stream.write(query)

    subprocess.run(
        [ZH_EXECUTABLE, 
         str(windowsize), str(minsize), str(maxsize), temp],
        check=True, stdout=PIPE, stderr=DEVNULL,
        input=query, encoding='ascii'
    )
    with open(temp + ".Z-SCORE", 'r') as stream:
        df = pd.read_csv(stream,
                         names=['Start', 'End', 'nu-1', 'nu-2', 'nu-3', 
                                'ZH-Score', 'Sequence', 'Conformation'],
                         skiprows=1, sep='\s+')
    os.remove(temp)
    os.remove(temp + ".Z-SCORE")
    return df[['Start', 'End', 'ZH-Score', 'Sequence', 'Conformation']]
```

```python
def get_genomes(file_name):
  df = pd.DataFrame(columns=['Start', 'End', 'Strand'])
  with open(f'{file_name}_genomic.gff') as file:
    for line in file:
      tokens = line.split('\t')
      if len(tokens) == 9 and tokens[2] == 'gene':
        df = df.append({'Start': int(tokens[3]), 'End': int(tokens[4]), 'Strand': tokens[6]}, ignore_index=True)
  return df

lengths = []
for name in file_names:
  df = get_genomes(name)
  df.to_csv(f'{name}_genomic.csv', index=False)
  lengths.append(df['End'].subtract(df['Start']).sum())
  print(f'{name} -> genes number: {len(df.index)}')
```

```python
from Bio import SeqIO
import matplotlib.pyplot as plt

index_l = 0
for name in file_names:
  with open(name + "_genomic.fna", 'r') as stream:
    sequences = list(SeqIO.parse(stream, format='fasta'))
  for s in sequences:
    print(f"{s.description} \n\t=> {len(s)}")
  seq = sequences[0]
  df = zhunt(str(seq.seq))
  if len(sequences) > 1:
    dataframes = []
    dataframes.append(df)
    seq2 = sequences[1]
    df2 = zhunt(str(seq2.seq))
    dataframes.append(df2)
    df = pd.concat(dataframes, axis=0)
  print(f'z-dnas total: {len(df)}')
  df = df[df['ZH-Score'] > 500]
  df = df.sort_values(by='ZH-Score')
  plt.hist(df['ZH-Score'], bins = 100)
  plt.title(f'{organisms_names[name]}')
  plt.xlabel('Score', fontsize=12)
  plt.ylabel('Volume', fontsize=12)
  plt.show()
  length = lengths[index_l]
  index_l += 1
  z_dna_length = df['End'].subtract(df['Start']).sum()
  print(f'{name}_genomic -> total z-dnas : {len(df)} total z-dnas length: {z_dna_length} sum length: {length}')
  with open(name + ".zhunt.bed5", 'w') as zhunt_bed:
    with open(name + "_genomic.zhunt.txt", 'w') as zhunt_f:
      index_z_dna = 1
      df = df.sort_values(by=['Start'])
      for index, row in df.iterrows():
        seq_start = row['Start']
        seq_end = row['End']
        seq_score = row['ZH-Score']
        zhunt_f.write(f'{seq_start}\t{seq_end}\t{seq_score}\n')
        z_dna_name = f'{name}_z_dna_{index_z_dna}'
        zhunt_bed.write(f'chr1\t{seq_start}\t{seq_end}\t{z_dna_name}\t{seq_score}\n')
        index_z_dna += 1
```

```python
import pandas as pd

def count(filename):
  non_blank_count = 0
  with open(filename) as infp:
    for line in infp:
       if line.strip():
          non_blank_count += 1
  return non_blank_count

organisms_names = {
    'GCF_003285265.1_ASM328526v1' : 'Paracoccus mutanolyticus',
    'GCF_008824185.1_ASM882418v1': 'Paracoccus pantotrophus',
    'GCF_004683865.2_ASM468386v2' : 'Paracoccus liaowanqingii',
    'GCF_003324675.1_ASM332467v1' : 'Paracoccus suum',
    'GCF_002749495.1_ASM274949v1' : 'Paracoccus yeei'
}

df = pd.DataFrame(columns=['Name' ,'Promoter', 'Gene', 'Non-gene'])
for f_name in file_names:
  name = organisms_names[f_name]
  promoter = count(f'{f_name}_tss_zhunt.bed')
  gene = count(f'{f_name}_gene_zhunt.bed')
  non_gene = count(f'{f_name}_none_zhunt.bed')
  total = promoter + gene + non_gene
  df = df.append({'Name': name, 'Promoter': 100 * promoter / total, 'Gene': 100 * gene / total, 'Non-gene' : 100 * non_gene / total}, ignore_index=True)

df.plot(
    x = 'Name',
    kind = 'barh',
    stacked = True,
    title = 'Z-DNA distribution (%)',
    mark_right = True
)
```

```python
def prot_clusters_2_df(file_name):
  name1 = 'GCF_002749495.1_ASM274949v1'
  name2 = 'GCF_003285265.1_ASM328526v1'
  name3 = 'GCF_003324675.1_ASM332467v1'
  name4 = 'GCF_004683865.2_ASM468386v2'
  name5 = 'GCF_008824185.1_ASM882418v1'
  df = pd.DataFrame(columns=['Genes',name1, name2, name3, name4, name5, 'Z-DNAs', 'Z-DNAs_in_promoter'])
  with open(file_name) as file:
    for line in file:
      tokens = line.split()
      if len(tokens) == 8 and tokens[1] != 'Genes':
        df = df.append({'Genes': int(tokens[1]), name1: tokens[3], name2: tokens[4], name3: tokens[5], name4: tokens[6], name5: tokens[7], 'Z-DNAs': 0, 'Z-DNAs_in_promoter': 0}, ignore_index=True)
  return df
```

```python
def list_of_seqs(file_name_end, size = 4):
  dfs= {}
  for name in file_names:
    with open(f'{name}{file_name_end}') as file:
      if size == 4:
        df = pd.DataFrame(columns=['Name', 'Start', 'End'])
      elif size == 5:
        df = pd.DataFrame(columns=['Name', 'Start', 'End', 'Score'])
      else:
        df = pd.DataFrame(columns=['Name', 'Start', 'End', 'Strand'])
      for line in file:
        tokens = line.split()
        if len(tokens) == size:
          if size == 4:
            df = df.append({'Name': tokens[3], 'Start': tokens[1], 'End': tokens[2]}, ignore_index=True)
          elif size == 5:
            df = df.append({'Name': tokens[3], 'Start': tokens[1], 'End': tokens[2], 'Score': tokens[4]}, ignore_index=True)
          else:
            df = df.append({'Name': tokens[3], 'Start': tokens[1], 'End': tokens[2], 'Strand': tokens[5]}, ignore_index=True)
      dfs[name] = df
  return dfs
```

```python
in_gene = list_of_seqs('_all_zhunt.bed')
in_tss = list_of_seqs('_tss_zhunt.bed')

for i, row in df_clusters.iterrows():
    z_dnas_total = 0
    z_dnas_in_ptomoter_total = 0
    for name in file_names:
      if df_clusters.at[i,name] != '*':
        df_in_gene = in_gene[name]
        df_in_tss = in_tss[name]
        gene_names = df_clusters.at[i,name].split(',')
        for gene_name in gene_names:
          if gene_name in df_in_gene['Name'].values:
            z_dnas_total += len(df_in_gene[df_in_gene['Name'] == gene_name])
          if gene_name in df_in_tss['Name'].values:
            z_dnas_in_ptomoter_total += 1
    df_clusters.at[i,'Z-DNAs'] = z_dnas_total
    df_clusters.at[i,'Z-DNAs_in_promoter'] = z_dnas_in_ptomoter_total
```
